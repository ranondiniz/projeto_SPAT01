```python
# Converte a coluna de data e hora para o tipo datetime
detran_DF['data_hora_boletim'] = pd.to_datetime(detran_DF['data_hora_boletim'])

```

```python
# Cria uma função para classificar o turno com base na hora
def classificar_turno(hora):
    if hora.hour >= 0 and hora.hour < 12:
        return 'Diurno'
    elif hora.hour >= 12 and hora.hour < 18:
        return 'Vespertino'
    else:
        return 'Noturno'
```

```python
# Aplica a função à coluna de data e hora para criar a nova coluna "turno"
detran_DF['turno'] = detran_DF['data_hora_boletim'].apply(classificar_turno)

# Exibe o DataFrame resultante
detran_DF.head()
```

```python
# Amostragem Estratificada

tamanho_amostra = 200
amostras_por_classe = []
classes = DF_detran['cod_severidade'].unique()

qtde_por_classe = round(tamanho_amostra / len(classes))

for indice in classes:
  lista_indices = DF_detran['cod_severidade'] == indice
  dados_classes = DF_detran[lista_indices]
  amostra_classe = dados_classes.sample(qtde_por_classe, replace=True)
  amostras_por_classe.append(amostra_classe)

amostra_DF = pd.concat(amostras_por_classe)
amostra_DF.head(200)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=amostra_DF, x='cod_severidade', )
plt.show()
```

Sendo o gráfico acima um gráfico de barras (contagem), é possível observar que a operação de obtenção de amostras foi bem-sucedida, visto que os valores da coluna 'cod_severidade' apresentam a mesma proporção no volume dos dados.

```python
# Lista de colunas categóricas

colunas_categoricas = ['condutor', 'cod_severidade', 'desc_severidade', 'sexo', 'cinto_seguranca', 'Embreagues', 'categoria_habilitacao', 'especie_veiculo', 'pedestre', 'passageiro', 'turno']

# Loop para calcular frequência e moda para colunas categóricas
for col in colunas_categoricas:
    frequencia = amostra_DF[col].value_counts()
    moda = amostra_DF[col].mode()
    print(f"Frequência para {col}:\n{frequencia}\n")
    print(f"Moda para {col}:\n{moda}\n")
```


O código acima tem a função de criar uma lista de colunas categóricas, e essas colunas que serão identificadas serão classificadas como colunas_categoricas. Após essa primeira etapa, o código realiza um loop para calcular a frequência e a moda de cada uma dessas colunas no DataFrame amostra_DF. O resultado do loop, mostrará a frequência de cada categoria em cada coluna categórica e a moda (valor mais frequente) para cada coluna. A frequência informa quantas vezes cada categoria aparece nos dados, enquanto a moda identifica a categoria que é a mais comum, como por exemplo, a frequência para sexo masculino é de 146, e a frequência para categoria_habilitação seria N com 96.

```python
import matplotlib.pyplot as plt


colunas_categoricas = ['condutor', 'cod_severidade', 'desc_severidade', 'sexo', 'cinto_seguranca', 'Embreagues', 'categoria_habilitacao', 'especie_veiculo', 'pedestre', 'passageiro', 'turno']


for col in colunas_categoricas:
    # Calcule a frequência das categorias
    frequencia = amostra_DF[col].value_counts()

    # Plote o gráfico de contagem
    plt.figure(figsize=(8, 6))
    sns.countplot(x=col, data=amostra_DF)
    plt.xlabel(col)
    plt.ylabel('Contagem')
    plt.title(f'Gráfico de Contagem para {col}')
    print('')

    # Gire o eixo x em 90 graus
    plt.xticks(rotation=45)

    plt.show()

    print('')

    # Plote o gráfico de pizza
    plt.figure(figsize=(6, 6))
    tamanhos = amostra_DF[col].value_counts().values
    categorias = amostra_DF[col].value_counts().index
    plt.pie(tamanhos, labels=categorias, autopct='%1.1f%%')
    plt.title(f'Gráfico de Pizza para {col}')
    plt.show()

    print('')

```

```python
# Verificando outiliers das colunas numéricas
# Há apenas 3 colunas com dados numéricos

colunas = amostra_DF.columns
for coluna in colunas:
    # Verifica se a coluna contém dados numéricos antes de criar o boxplot
    if amostra_DF[coluna].dtype in ['int64', 'float64']:
        # Cria um boxplot para a coluna atual
        plt.figure(figsize=(8, 6))  # Define o tamanho da figura
        sns.boxplot(x=amostra_DF[coluna])
        plt.title(f'Boxplot da Coluna {coluna}')  # Adiciona título ao gráfico
        plt.show()  # Mostra o gráfico
```

Tendo em vista que os gráficos acima exibem os outliers das colunas númericas sendo elas respectivamente: Nº_envolvidos, cod_severidade e idade. É possivel inferir que:
* A única coluna que não apresenta outliers é a do cod_severidade, ou seja, todos os valores da coluna estão dentro de um intervalo razoável, não havendo desse modo observações que sejam muito diferentes do restante do conjunto de dados.
* Os outliers da coluna Nº_envolvido, bem como os da coluna idade, apresentam uma certa tendência em seus valores, fazendo com que tenhamos mais atenção a natureza desses dados pois isso pode indicar informações importantes sobre os dados ou até mesmo erros sistemáticos na coleta ou entrada de dados.

```python
coluna_numerica = 'Nº_envolvido'

# Calcula o IQR para a coluna
Q1 = amostra_DF[coluna_numerica].quantile(0.25)
Q3 = amostra_DF[coluna_numerica].quantile(0.75)
IQR = Q3 - Q1

# Define os limites para outliers (1.5 vezes o IQR abaixo de Q1 e acima de Q3)
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Remove outliers com base nos limites
amostra_sem_outliers = amostra_DF[(amostra_DF[coluna_numerica] >= limite_inferior) & (amostra_DF[coluna_numerica] <= limite_superior)]

# O DataFrame 'amostra_sem_outliers' agora contém apenas os valores que não são considerados outliers no boxplot.
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x=amostra_sem_outliers['Nº_envolvido'])
plt.title(f'Boxplot da Coluna Nº_envolvido')
plt.show()
```

Por meio do gráfico podemos notar a distribuição estatística da coluna Nº_envolvido. Geralmente usa-se esse gráfico, para visualizar a dispersão e a tendência central dos dados, bem como identificar possíveis outliers (valores que estão significativamente afastados dos valores centrais), neste gráfico gerado não são encontrados outiers.
